```python
import numpy as np
import matplotlib.pyplot as plt

from astroscrappy import detect_cosmics

%matplotlib widget
```

```python
data = np.load('/Users/zgl12/Modules/AstroColour/temp_data_bkging.npy')
```

```python
plt.figure()
plt.imshow(data, origin='lower', vmin = np.nanpercentile(data, 1), vmax = np.nanpercentile(data, 99), cmap= 'gray')
plt.show()
```

```python

crmask, cleaned = detect_cosmics(data, 
                                  sigclip=4.5,       # sigma threshold for detection
                                  sigfrac=0.3,        # neighboring pixels fraction
                                  objlim=5.0,         # contrast between CR and object
                                  gain=3.0,           # if not gain-corrected, set appropriately
                                  readnoise=5.0,      # your camera's read noise (in e-)
                                  satlevel=65535.0,   # saturation limit (to avoid false positives)
                                  niter=4,            # number of iterations
                                  cleantype='medmask' # how to interpolate over CRs
                                  )
```

```python
plt.figure()
plt.imshow(cleaned, origin='lower', vmin = np.nanpercentile(cleaned, 1), vmax = np.nanpercentile(cleaned, 99), cmap= 'gray')
plt.show()
```
